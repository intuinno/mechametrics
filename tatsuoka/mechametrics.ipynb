```python
import pandas as pd
import numpy as np
import re
import torch
```

```python
df = pd.read_csv('data/tatsuoka.csv')
df.head(22895)
```

```python
questions = df['question'].tolist()
```

```python
idx = [3,5,6,7,9,11,12]
search = r'What is \$(\\displaystyle )?(\{(\d+)\})?\s*(\\d?frac{(\d+)}{(\d+)})?\s?([\-\+])\s?(\{(\d+)\})?(\\d?frac{(\d+)}{(\d+)})?\$\?'
question_num = [] 
for i, q in enumerate(questions):
#     print i
    match = re.search(search, q)
    if match:
        question_num.append([match.group(i) for i in idx])
    else:
        print "Something wrong",i, q
```

```python
q_df = pd.DataFrame(question_num)
```

```python
q_df.to_csv('questions.csv', na_rep=0)
```
